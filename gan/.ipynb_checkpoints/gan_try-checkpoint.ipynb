```python
import numpy as np
import tensorflow as tf
from IPython import display
```

```python
mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
def ima_pixel(x):
    ima_size=x.shape[1]*x.shape[2]
    return ima_size
```

```python
ima_size=ima_pixel(x_train)
noise_size=100
batch_size=100
```

```python
def noise(n_rows,n_cols):
    return np.random.normal(size=(n_rows, n_cols))

def xavier_init(size):
    in_dim = size[0] if len(size) == 1 else size[1]
    stddev = 1. / np.sqrt(float(in_dim))
    return tf.random_uniform(shape=size, minval=-stddev, maxval=stddev)

def images_to_vectors(images):
    return images.reshape(images.shape[0], ima_size)

def vectors_to_images(vectors):
    return vectors.reshape(vectors.shape[0], 28, 28, 1)
```

```python
tf.reset_default_graph()
```

```python
with tf.name_scope('Input'):
    X = tf.placeholder(tf.float32, shape=[None, ima_size], name='X')
    Z = tf.placeholder(tf.float32, shape=[None, noise_size], name='Z')
```

```python
def weight_variable(shape):
    W = tf.get_variable("W", shape=shape, initializer=tf.contrib.layers.xavier_initializer())
    return W

def bias_variable(shape):
    B = tf.get_variable('b',shape=shape, initializer=tf.contrib.layers.xavier_initializer())
    return B
```

```python
def layer(x, name, shape,leaky=True, drop_out=True):
    with tf.variable_scope(name,reuse=True):
        
        W = weight_variable(shape=shape)
        tf.summary.histogram('weight', W)
        
        B = bias_variable(shape=shape[1])
        tf.summary.histogram('bias', B)
        
        layer = tf.matmul(x, W)+B
        if leaky==True:
            layer = tf.nn.leaky_relu(layer, .2)
        if drop_out==True:
            layer = tf.nn.dropout(layer, .3)
        
        return layer
```

## Discriminator

```python
def discriminator(x):
    d1 = layer(x, name='d1', shape=[ima_size, 1024], leaky=True, drop_out=True)
    d2 = layer(d1, name='d2', shape=[1024, 512], leaky=True, drop_out=True)
    d3 = layer(d2, name='d3', shape=[512, 256], leaky=True, drop_out=True)
    d4 = layer(d3, name='d4', shape=[256, 1], leaky=False, drop_out=False)
    return d4
```

```python
def generator(z):
    g1 = layer(z, name='g1', shape=[noise_size, 256], leaky=True, drop_out=False)
    g2 = layer(g1, name='g2', shape=[256, 512], leaky=True, drop_out=False)
    g3 = layer(g2, name='g3', shape=[512, 1024], leaky=True, drop_out=False)
    g4 = layer(g3, name='g4', shape=[1024, 784], leaky=False, drop_out=False)
    return g4
```

```python
G_sample = generator(Z)
D_real = discriminator(X)
```

```python
D_fake = discriminator(G_sample)
```

```python
# Losses
with tf.variable_scope('train'):
    with tf.variable_scope('loss'):
        D_loss_real = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=D_real, labels=tf.ones_like(D_real)))
        D_loss_fake = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=D_fake, labels=tf.zeros_like(D_fake)))
        D_loss = D_loss_real + D_loss_fake
        G_loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=D_fake, labels=tf.ones_like(D_fake)))
    with tf.variable_scope('Optimizer'):
        # Optimizers
        D_opt = tf.train.AdamOptimizer(2e-4).minimize(D_loss)
        G_opt = tf.train.AdamOptimizer(2e-4).minimize(G_loss)
```

# Train

```python
num_test_samples = 16
test_noise = noise(num_test_samples, noise_size)
```

### inits

```python
def get_next_batch(x, start, end):
    x_batch = x[start:end]
    #y_batch = y[start:end]
    return x_batch #y_batch
```

```python
# Initialize the variables
init = tf.global_variables_initializer()
# Merge all summaries
merged = tf.summary.merge_all()
```

```python
epochs = 10
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
def plot(samples):
    fig = plt.figure(figsize=(2, 2))
    gs = gridspec.GridSpec(4, 4)
    gs.update(wspace=0.05, hspace=0.05)

    for i, sample in enumerate(samples):
        ax = plt.subplot(gs[i])
        plt.axis('off')
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        ax.set_aspect('equal')
        plt.imshow(sample.reshape(28, 28), cmap='Greys_r')

    return fig
```

```python
# Start interactive session
sess = tf.InteractiveSession()
sess.run(init)
global_step=0
summary_writer = tf.summary.FileWriter('logs', sess.graph)
saver = tf.train.Saver()

num_tr_iter = int(len(y_train) / batch_size)
for epoch in range(epochs):
    print('Training epoch: {}'.format(epoch + 1))
    
    for iteration in range(num_tr_iter):
        # 1. Train Discriminator
        global_step += 1
        start = iteration * batch_size
        end = (iteration + 1) * batch_size
        x_batch = get_next_batch(x_train, start, end)

        X_batch = images_to_vectors(x_batch)
        feed_dict = {X: X_batch, Z: noise(batch_size, noise_size)}
        _, d_error, d_pred_real, d_pred_fake = sess.run(
            [D_opt, D_loss, D_real, D_fake], feed_dict=feed_dict
        )

        # 2. Train Generator
        feed_dict = {Z: noise(batch_size, noise_size)}
        _, g_error = sess.run(
            [G_opt, G_loss], feed_dict=feed_dict
        )
        
        if iteration % 100 == 0:
            #display.clear_output(True)
            # Generate images from test noise
            test_images = sess.run(
                G_sample, feed_dict={Z: test_noise}
            )
            
            test_images = vectors_to_images(test_images)
            plot(test_images)
            print("iter {0}:\td_error={1},\tg_error={2}".
                  format(iteration, d_error,g_error))
                
            
```
